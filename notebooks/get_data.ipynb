```python
import requests
import pandas as pd
import os
```

```python
endpoint = 'https://data.lacity.org/resource/2nrs-mtv8.json'
```

```python
response = requests.get(url=endpoint,
             headers={"Accept":"application/json",
                      "X-App-Token":os.getenv('chicago_app_token')},
             params={'$where':"date_occ between '2022-01-10T00:00:00' and '2022-12-31T23:59:59'"},
             auth=(os.getenv('chicago_key_id'),os.getenv('chicago_key_secret'))
)
```

```python
data = pd.json_normalize(data=response.json())
```
